```python
import csv
A = []
B = []
with open('Corp_Summary.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:
        A.append(' '.join(row))
    print(A)
    for i in range(len(A)):
        B.append(A[i].split(';'))
    print(B)
```

```python
def depart_teams(B):
    '''
    Данная функция получает на вход массив массивов, в котором записаны все данные. 
    Мы проходим по всем данным и собираем информацию о существующих департаментах и работающих в них командах.
    В выводе будет словарь, в котором ключами являются 
    '''
    C = {}
    for i in range(len(B)):
        if C.get(B[i][1])==None:
            C[B[i][1]] = set()
            C[B[i][1]].add(B[i][2])
        else:
            C[B[i][1]].add(B[i][2])
    print(C)
```

```python
def wages(B):
    '''
    Данная функция принимает на вход массив массивов, в котором записаны все данные.
    В результате работы функции создается запись, в которой содержатся минимальная, максимальная и средняя зарплаты по каждому департаменту.
    '''
    D = {}
    for i in range(len(B)):
        if B[i][1]!= 'Департамент':
            if D.get(B[i][1])==None:
                D[B[i][1]] = []
                D[B[i][1]].append(int(B[i][-1]))
            else:
                D[B[i][1]].append(int(B[i][-1]))
    for j in D:
        print(j, ':', 'численность:', len(D[j]), '; минимальная заработная плата:', min(D[j]), '; максимальная заработная плата:', max(D[j]), '; средняя заработная плата:', sum(D[j])/len(D[j]), '.')
```

```python
def wages_1(B):
    '''
    Данная функция принимает на вход массив массивов, в котором записаны все данные.
    Функция возвращает словарь с зарплатами по департаментам.
    '''
    D = {}
    for i in range(len(B)):
        if B[i][1]!= 'Департамент':
            if D.get(B[i][1])==None:
                D[B[i][1]] = []
                D[B[i][1]].append(int(B[i][-1]))
            else:
                D[B[i][1]].append(int(B[i][-1]))
    return D
```

```python
def summary(D):
    '''
    Данная функция принимает на вход словарь, содержащий нужные данные о зарплате.
    Функция делает запись в файле csv.
    '''
    with open('Corp_Summary(1).csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, delimiter=' ',
                                quotechar='|', quoting=csv.QUOTE_MINIMAL)
        for j in D:
            a = j + ':' + 'минимальная заработная плата:' + str(min(D[j])) + '; максимальная заработная плата:' + str(max(D[j])) + '; средняя заработная плата:' + str(sum(D[j])/len(D[j])) + '.'
            writer.writerow(a)
```
